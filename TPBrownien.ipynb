###Question 1 : Trajectoire du mouvement brownien standard


Le mouvement brownien standard est un processus stochastique avec des accroissements indépendants, stationnaires et gaussiens, et dont les trajectoires sont continues presque sûrement. La trajectoire d’un mouvement brownien standard B_t sur l’intervalle [0, T] peut être simulée en générant des incréments gaussiens et en cumulant ces incréments.

```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion_standard(T, steps=1000):
    dt = T / steps
    increments = np.random.normal(0, np.sqrt(dt), steps)
    path = np.cumsum(increments)
    return np.concatenate(([0], path))

def plot_brownian_motion_standard(T, steps=1000):
    path = brownian_motion_standard(T, steps)
    time = np.linspace(0, T, steps + 1)
    plt.plot(time, path)
    plt.xlabel('Time')
    plt.ylabel('Position')
    plt.title('Standard Brownian Motion')
    plt.show()

# Trajectoire du mouvement brownien standard sur [0, T]
plot_brownian_motion_standard(1)
```

Cette figure montre la trajectoire d'un mouvement brownien standard sur l'intervalle $[0, 1]$. Les incréments sont gaussiens avec une moyenne de 0 et une variance proportionnelle au temps Δt. On observe les fluctuations aléatoires autour de la position initiale (0).

###Question 2 : Trajectoire du mouvement brownien avec drift et variance


Un mouvement brownien avec drift $\mu$ et variance $\sigma^2$ est un processus $X_t$ défini par $X_t = B_t + \mu$ t où $B_t$ est un mouvement brownien standard. Les incréments sont alors générés avec une moyenne $\mu \Delta t$ et une variance $\sigma^2 \Delta t$.

```python
def brownian_motion(T, steps=1000, mu=0, sigma=1):
    dt = T / steps
    increments = np.random.normal(mu * dt, sigma * np.sqrt(dt), steps)
    path = np.cumsum(increments)
    return np.concatenate(([0], path))

def plot_brownian_motion(T, mu=0, sigma=1, steps=1000):
    path = brownian_motion(T, steps, mu, sigma)
    time = np.linspace(0, T, steps + 1)
    plt.plot(time, path)
    plt.xlabel('Time')
    plt.ylabel('Position')
    plt.title(f'Brownian Motion with mu={mu}, sigma={sigma}')
    plt.show()

# Trajectoire du mouvement brownien avec drift mu et variance sigma^2
plot_brownian_motion(1, mu=0.1, sigma=0.5)
```

Cette figure montre la trajectoire d’un mouvement brownien avec un drift $\mu = 0.1$ et une variance $\sigma^2 = 0.25$ sur l’intervalle [0, 1]. On observe que la trajectoire tend à s’élever légèrement en raison du drift positif tout en présentant des fluctuations dues à la variance.

###Question 3 : Trajectoire d’un mouvement brownien bidimensionnel


Un mouvement brownien bidimensionnel est un processus où chaque composante suit un mouvement brownien indépendant. Cela signifie que pour chaque composante, les incréments sont générés indépendamment.

```python
def brownian_motion_2D(T, steps=1000, mu=[0, 0], sigma=[1, 1]):
    dt = T / steps
    increments_x = np.random.normal(mu[0] * dt, sigma[0] * np.sqrt(dt), steps)
    increments_y = np.random.normal(mu[1] * dt, sigma[1] * np.sqrt(dt), steps)
    path_x = np.cumsum(increments_x)
    path_y = np.cumsum(increments_y)
    return np.concatenate(([0], path_x)), np.concatenate(([0], path_y))

def plot_brownian_motion_2D(T, mu=[0, 0], sigma=[1, 1], steps=1000):
    path_x, path_y = brownian_motion_2D(T, steps, mu, sigma)
    plt.plot(path_x, path_y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'2D Brownian Motion with mu={mu}, sigma={sigma}')
    plt.show()

# Trajectoire du mouvement brownien bidimensionnel
plot_brownian_motion_2D(1, mu=[0.1, 0.1], sigma=[0.5, 0.5])

plot_brownian_motion_2D(1, mu=[0.1, 0.1], sigma=[0.5, 0])

plot_brownian_motion_2D(1, mu=[0.1, 0.1], sigma=[0, 0.5])
```

Cette figure montre la trajectoire d’un mouvement brownien bidimensionnel avec des paramètres de drift $\mu = [0.1, 0.1]$ et de variance $\sigma = [0.5, 0.5]$ sur l’intervalle [0, 1]. Chaque composante suit un mouvement brownien indépendant, ce qui donne une trajectoire dans le plan 2D.
En assignant σ à 0 sur les autres figures on obtient  deux mouvements brownien avec drifts et variances monodimensionnels

###Question 4 : Trajectoire d’un mouvement brownien tridimensionnel


Un mouvement brownien tridimensionnel suit le même principe que le mouvement brownien bidimensionnel, avec une composante supplémentaire.

```python
def brownian_motion_3D(T, steps=1000, mu=[0, 0, 0], sigma=[1, 1, 1]):
    dt = T / steps
    increments_x = np.random.normal(mu[0] * dt, sigma[0] * np.sqrt(dt), steps)
    increments_y = np.random.normal(mu[1] * dt, sigma[1] * np.sqrt(dt), steps)
    increments_z = np.random.normal(mu[2] * dt, sigma[2] * np.sqrt(dt), steps)
    path_x = np.cumsum(increments_x)
    path_y = np.cumsum(increments_y)
    path_z = np.cumsum(increments_z)
    return np.concatenate(([0], path_x)), np.concatenate(([0], path_y)), np.concatenate(([0], path_z))

def plot_brownian_motion_3D(T, mu=[0, 0, 0], sigma=[1, 1, 1], steps=1000):
    from mpl_toolkits.mplot3d import Axes3D
    path_x, path_y, path_z = brownian_motion_3D(T, steps, mu, sigma)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(path_x, path_y, path_z)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title(f'3D Brownian Motion with mu={mu}, sigma={sigma}')
    plt.show()

# Trajectoire du mouvement brownien tridimensionnel
plot_brownian_motion_3D(1, mu=[0.1, 0.1, 0.1], sigma=[0.5, 0.5, 0.5])

```

Cette figure montre la trajectoire d'un mouvement brownien tridimensionnel avec des paramètres de drift $\mu = [0.1, 0.1, 0.1]$ et de variance $\sigma = [0.5, 0.5, 0.5]$ sur l'intervalle \([0, 1]\). Chaque composante suit un mouvement brownien indépendant, ce qui donne une trajectoire dans l'espace 3D.

#Question 1 : Principe de réflexion


Le principe de réflexion pour le mouvement brownien standard indique que la probabilité que le maximum du mouvement brownien dépasse une certaine valeur a est deux fois la probabilité que le mouvement brownien atteigne a à un temps donné.


```python
def reflection_principle(T, a, trials=10000):
    sup_B = np.zeros(trials)
    B_t = np.random.normal(0, np.sqrt(T), trials)
    for i in range(trials):
        path = np.cumsum(np.random.normal(0, np.sqrt(T/trials), trials))
        sup_B[i] = np.max(path)
    return np.mean(sup_B >= a), np.mean(np.abs(B_t) >= a),np.mean(B_t>=a)

T = 1
a = 1
P_sup, P_abs ,Pbt = reflection_principle(T, a)
print(f"Estimated P(sup B_s >= {a}): {P_sup}")
print(f"Estimated P(|B_t| >= {a}): {P_abs}")

print(f"Estimated P(B_t >= {a}): {Pbt}")
```

###Question 2 : Distribution de $Y = \arg\max_{s \in [0,1]} B_s$


Nous générons un échantillon de valeurs de Y et traçons un histogramme pour comparer avec la densité de la distribution arcsinus.

```python
def argmax_brownian(T=1, K=10000, steps=1000):
    max_indices = np.zeros(K)
    for i in range(K):
        path = np.cumsum(np.random.normal(0, np.sqrt(T/steps), steps))
        max_indices[i] = np.argmax(path)
    return max_indices / steps

Y_samples = argmax_brownian()
plt.hist(Y_samples, bins=50, density=True, alpha=0.6, color='g', label='Histogram of Y samples')

# Arcsine distribution
x = np.linspace(0, 1, 1000)
f_x = 1 / (np.pi * np.sqrt(x * (1 - x)))
plt.plot(x, f_x, 'r', label='Arcsine distribution')
plt.xlabel('Y')
plt.ylabel('Density')
plt.legend()
plt.title('Histogram of Y and Arcsine distribution')
plt.show()
```

on observe ici que Y suit la distribution d'arcsinus . Le résultat est donc cohérent

###Question 3 : Vérification numérique de  $P(T = T_b) = \frac{a}{a + b} $



On se donne a > 0 et b > 0. Notons :

$	•	T_{-a} := \inf\{ t > 0 : B_t = -a \}$

$	•	T_b := \inf\{ t > 0 : B_t = b \}$
$	•	T := T_{-a} \wedge T_b $

L’affirmation à vérifier est que  $P(T = T_b) = \frac{a}{a + b} $.

```python
import numpy as np

def simulate_hitting_times(a, b, T=1, steps=10000, trials=10000):
    dt = T / steps
    hit_b = 0

    for _ in range(trials):
        B_t = np.cumsum(np.random.normal(0, np.sqrt(dt), steps))

        T_a = np.inf
        T_b = np.inf

        for i in range(steps):
            if B_t[i] <= -a:
                T_a = i * dt
                break
            elif B_t[i] >= b:
                T_b = i * dt
                break

        if T_b < T_a:
            hit_b += 1

    prob_hit_b = hit_b / trials
    theoretical_prob = a / (a + b)

    return prob_hit_b, theoretical_prob

# Paramètres
a = 1
b = 2
simulated_prob, theoretical_prob = simulate_hitting_times(a, b)
print(f"Simulated P(T = T_b): {simulated_prob}")
print(f"Theoretical P(T = T_b): {theoretical_prob}")
```

## Exercice 3 : Autour du Théorème de Donsker

### Question 1 : Tracer les trajectoires de $(S_n)_n$ jusqu'à un instant $N$.


Pour illustrer le Théorème de Donsker, nous devons simuler une marche aléatoire discrète $(S_n)_n$ définie par $S_n = \sum_{k=1}^n X_k$ où $X_k$ sont des variables aléatoires i.i.d. centrées et réduites.

```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(N):
    X = np.random.normal(0, 1, N)
    S = np.cumsum(X)
    return S

def plot_random_walk(N):
    S = random_walk(N)
    time = np.arange(1, N+1)
    plt.plot(time, S)
    plt.xlabel('Steps')
    plt.ylabel('S_n')
    plt.title('Random Walk (S_n)')
    plt.show()

# Tracer les trajectoires de (S_n)_n jusqu'à un instant N
plot_random_walk(1000)
```

### Question 2 : Tracer le graphe de $C_n$ pour un $n$ fixé.

La variable aléatoire $C_n(t)$ est définie par une interpolation et une normalisation des trajectoires de la marche aléatoire discrète. Nous traçons  cette fonction pour un $n$ fixé .

```python
import numpy as np
import matplotlib.pyplot as plt

def interpolate_random_walk(N, steps):
    X = np.random.normal(0, 1, N)
    S = np.cumsum(X)
    t = np.linspace(0, 1, steps)
    Cn_t = np.zeros(steps)
    for i in range(steps):
        k = int(t[i] * N)
        if k < N-1:
            Cn_t[i] = 1/np.sqrt(N) * (S[k] + (t[i]*N - k) * X[k+1])
        else:
            Cn_t[i] = 1/np.sqrt(N) * S[-1]
    return t, Cn_t

def plot_interpolated_random_walk(N, steps=10000):
    t, Cn_t = interpolate_random_walk(N, steps)
    plt.plot(t, Cn_t)
    plt.xlabel('t')
    plt.ylabel('C_n(t)')
    plt.title(f'Interpolated Random Walk C_n(t) for n={N}')
    plt.show()

# Tracer le graphe de X_n(t) pour un n fixé
plot_interpolated_random_walk(1000)
```

###Question 3 :Illustrer le résultat de Donsker en simulant $Cn$ pour $n$ suffisament grand.
Le Théorème de Donsker stipule que pour $n$ suffisamment grand, $C_n(t)$ converge en loi vers un mouvement brownien standard. Nous allons simuler $C_n$ pour une grande valeur de $n$ et comparer avec une trajectoire de mouvement brownien standard.

```python
def plot_donsker_theorem(N, steps=10000):
    t, Xn_t = interpolate_random_walk(N, steps)
    B_t = np.cumsum(np.random.normal(0, np.sqrt(1/steps), steps))
    B_t = np.concatenate(([0], B_t))  # Initial value at 0
    t_B = np.linspace(0, 1, steps + 1)

    plt.plot(t, Xn_t, label='X_n(t)')
    plt.plot(t_B, B_t, label='Standard Brownian Motion')
    plt.xlabel('t')
    plt.ylabel('Value')
    plt.title(f'Convergence of X_n(t) to Brownian Motion for n={N}')
    plt.legend()
    plt.show()

# Illustrer le résultat de Donsker
plot_donsker_theorem(10000000)
```

### Question 1 : Mouvement brownien géométrique


Le mouvement brownien géométrique est souvent utilisé pour modéliser les prix d'actifs financiers. Il est défini par :
$ S_t = S_0 \exp \left( \left(\mu - \frac{\sigma^2}{2}\right) t + \sigma B_t \right) $



```python
def geometric_brownian_motion(T, S0, mu, sigma, steps=1000):
    dt = T / steps
    B_t = np.cumsum(np.random.normal(0, np.sqrt(dt), steps))
    t = np.linspace(0, T, steps)
    S_t = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * B_t)
    return t, S_t

def plot_geometric_brownian_motion(T, S0, mu, sigma, steps=1000):
    t, S_t = geometric_brownian_motion(T, S0, mu, sigma, steps)
    plt.plot(t, S_t)
    plt.xlabel('Time')
    plt.ylabel('S_t')

    plt.title(f'Geometric Brownian Motion with mu={mu}, sigma={sigma}')
    plt.show()

# Trajectoire du mouvement brownien géométrique
plot_geometric_brownian_motion(1, S0=100, mu=0.1, sigma=0.2)

plot_geometric_brownian_motion(1, S0=100, mu=0.005, sigma=0.1) #mu-sigma^2/2 =0
```

### Question 2 : Pont brownien

Le pont brownien est un mouvement brownien conditionné à être zéro aux temps 0 et 1.


```python
def brownian_bridge(T, steps=1000):
    dt = T / steps
    B_t = np.cumsum(np.random.normal(0, np.sqrt(dt), steps))
    t = np.linspace(0, T, steps)
    W_t = B_t - t / T * B_t[-1]
    return t, W_t

def plot_brownian_bridge(T, steps=1000):
    t, W_t = brownian_bridge(T, steps)
    plt.plot(t, W_t)
    plt.xlabel('Time')
    plt.ylabel('W_t')
    plt.title('Brownian Bridge')
    plt.show()

# Trajectoire du pont brownien
plot_brownian_bridge(1)
```

On retrouve  un mouvment ressemblant à un mouvement brownien commencant. en 0 et finissant en 0, on en conclut qu'on a bien tracé un mouvement brownien.

### Question 3 : Excursion brownienne


L'excursion brownienne est un pont brownien conditionné à rester positif. On utilise la transformée de Vervaat pour simuler ce processus.

```python
def brownian_excursion(T=1, steps=1000):
    dt = T / steps
    B_t = np.cumsum(np.random.normal(0, np.sqrt(dt), steps))
    T_min = np.argmin(B_t)
    W_t = np.concatenate((B_t[T_min:] - B_t[T_min], B_t[:T_min] - B_t[T_min]))
    W_t_excursion = np.abs(W_t)
    t = np.linspace(0, T, steps)
    return t, W_t_excursion

def plot_brownian_excursion(T=1, steps=1000):
    t, W_t_excursion = brownian_excursion(T, steps)
    plt.plot(t, W_t_excursion)
    plt.xlabel('Time')
    plt.ylabel('W_t')
    plt.title('Brownian Excursion')
    plt.show()

# Trajectoire de l'excursion brownienne
plot_brownian_excursion(1)
```

on retrouve ici un pont brownien positif donc on a bien tracé une
